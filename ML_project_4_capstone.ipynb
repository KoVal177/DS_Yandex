````python
def recovery(c, f, t):
    """
    –†–∞—Å—Å—á–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è
    """
    return c * (f - t) / (f * (c - t)) * 100

recovery_rough = data_train.apply(lambda row:
    recovery(
        row['rougher.output.concentrate_au'],
        row['rougher.input.feed_au'],
        row['rougher.output.tail_au']
    ),
    axis=1
)

recovery_final = data_train.apply(lambda row:
    recovery(
        row['final.output.concentrate_au'],
        row['rougher.input.feed_au'],
        row['final.output.tail_au']
    ),
    axis=1
)

print(mean_absolute_error(data_train['rougher.output.recovery'], recovery_rough))
print(mean_absolute_error(data_train['final.output.recovery'], recovery_final))
````

### –ü–æ–∏—Å–∫ –ª—É—á—à–∏—Ö –º–æ–¥–µ–ª–µ–π –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∞

````python
def score_smape(y, y_hat, **kwargs):
    '''
    –ú–µ—Ç—Ä–∏–∫–∞ sMAPE
    '''
    return sum([abs(a - b) / ((abs(a) + abs(b)) / 2) for a, b in zip(y, y_hat)]) / len(y)

def smape_total(smape_rougher, smape_final):
    '''
    –ú–µ—Ç—Ä–∏–∫–∞ –∏—Ç–æ–≥–æ–≤–∞–µ sMAPE
    '''
    return 0.25 * smape_rougher + 0.75 * smape_final
````

````python
def prepare_xy(features, targets, sample):
    '''
    –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –∏ –∑–∞–≤–∏—Å–∏–º—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö
    '''
    return (sample[features],
            sample[targets])
````

````python
def find_best_models(X, y, regressors, cv, scorer):
    '''
    –ü–µ—Ä–µ–±–æ—Ä –º–æ–¥–µ–ª–µ–π —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω–æ–π –º–µ—Ç—Ä–∏–∫–∏ –∏ –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
    '''
    best_models = []
    for name, estimator, parameters in regressors:
        print(name)

        model = GridSearchCV(
            estimator=estimator(),
            cv=cv,
            param_grid=parameters,
            scoring=scorer,
        )
        model.fit(X, y.values.ravel())
        best_models.append((name, model.best_estimator_, model.best_score_))
    return best_models
````

````python
def research_best_models(X, y, best_models):
    '''
    –°—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π
    '''
    for name, model, score in best_models:
        y_predict = model.predict(X)
        plot_df = pd.DataFrame(y.to_numpy().ravel(), (y.to_numpy().ravel() - y_predict.ravel())).reset_index()
        print(name)
        for metric_name, metric_score in (
            ('sMAPE score:', -score),
            ('r2 score:', r2_score(y, y_predict)),
            ('corr coef between error and y-actual:', plot_df.corr().iloc[0][0]),
        ):
            print(metric_name + '{:.1%}'.format(metric_score).ljust(10).rjust(50 - len(metric_name)))

        plt.scatter(x=y, y=y_predict, alpha=0.5)
        plt.title(name +' y-actual against y-predicted')
        plt.xlabel('y-actual')
        plt.ylabel('y-predicted')
        plt.xlim(0, 100)
        plt.ylim(0, 100)
        plt.show()

        plot_df.plot(x='index', y=0, kind='scatter', alpha=0.05);
        plt.title('Prediction error agains y-actual for ' + name)
        plt.xlabel('error')
        plt.ylabel('y-actual')
        plt.show()

        print()
````

–£ –º–æ–¥–µ–ª–µ–π –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–µ—Ä–µ–≤—å–≤ –ø—Ä–∏–Ω—Ü–∏–ø –ø—Ä–∏–Ω—è—Ç–∏—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–≤—è–∑–∞–Ω –Ω–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∞ –æ–Ω–∞ –ø—Ä–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏ –≤–æ–∑–¥–µ–π—Å—Ç–≤–∏—é –Ω–µ –ø–æ–¥–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

$y = \alpha_{st} + \beta_{st} \frac{X-mean}{std} = (\alpha_{st} - \frac{\beta_{st}mean}{std}) + (\frac{\beta_{st}}{std}) X$ - —Å–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–µ–π

````python
def compare_best_dummy(X, y, best_model, dummy_model):
    '''
    –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ —Å dummy
    '''
    total_score = []
    for name, model, score in (best_model, dummy_model):
        y_predict = model.predict(X)
        score = -score_smape(y.to_numpy().ravel(), y_predict.ravel())
        plot_df = pd.DataFrame(y.to_numpy().ravel(), (y.to_numpy().ravel() - y_predict.ravel())).reset_index()
        print(name)
        for metric_name, metric_score in (
            ('sMAPE score:', -score),
            ('r2 score:', r2_score(y, y_predict)),
            ('corr coef between error and y-actual:', plot_df.corr().iloc[0][0]),
        ):
            print(metric_name + '{:.1%}'.format(metric_score).ljust(10).rjust(50 - len(metric_name)))

        plt.scatter(x=y, y=y_predict, alpha=0.05)
        plt.title(name +' y-actual against y-predicted')
        plt.xlabel('y-actual')
        plt.ylabel('y-predicted')
        plt.xlim(0, 100)
        plt.ylim(0, 100)
        plt.show()

        plot_df.plot(x='index', y=0, kind='scatter', alpha=0.05);
        plt.title('Prediction error agains y-actual for ' + name)
        plt.xlabel('error')
        plt.ylabel('y-actual')
        plt.show()

        print()
        total_score.append((name, score))
        
    return total_score
````
