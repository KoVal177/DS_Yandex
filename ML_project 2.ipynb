–ü–æ—Å–∫–æ–ª—å–∫—É –¥–ª—è –ø–æ–∏—Å–∫–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π –º–æ–¥–µ–ª–∏ –ø–ª–∞–Ω–∏—Ä—É–µ—Ç—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å GridSearchCV, —Ç–æ –¥–∞–Ω–Ω—ã–µ —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –Ω–∞ –¥–≤–µ —á–∞—Å—Ç–∏ - —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é. –ó–∞—Ç–µ–º –æ–Ω–∏ –Ω–æ—Ä–º–∞–ª–∏–∑—É—é—Ç—Å—è –∏ –ø—Ä–æ–≤–æ–¥–∏—Ç—Å—è –æ–±—É—á–µ–Ω–∏–µ –Ω–∞ 3-–º–æ–¥–µ–ª—è—Ö:
* –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π;
* –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è;
* —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å;

```python
train_size = 0.8
test_size = 0.2


X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=test_size,
    random_state=RS,
    stratify=data[target_columns],
)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

# –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤ –±—É–¥—É—â–µ–º, –¥–æ –æ–±—Ä–∞–±–æ—Ç–∫–∏ Scaler
X_train_initial, y_train_initial = X_train, y_train
```

```python
scaler = StandardScaler().fit(X)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#### –î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

```python
parameters = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [2*n for n in range(1,10)],
    'max_features': ['auto', 'sqrt'],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
    'random_state': [RS],
}
estimator = DecisionTreeClassifier()
model = GridSearchCV(
    estimator=estimator,
    cv=10,
    param_grid=parameters,
    scoring='f1',
)

model.fit(X_train, y_train.values.ravel())
tree_cv_0 = model.best_estimator_
```

```python
print(model.best_estimator_)
print(model.best_params_)
```

```python
cv_score = cross_validate(
    model.best_estimator_,
    X_train,
    y_train.values.ravel(),
    cv=10,
    scoring=('f1','accuracy')
)
cv_score
```

```python
print('accuracy : {}'.format(cv_score['test_accuracy'].mean()))
print('f1 score (from cross_validate function): {}'.format(cv_score['test_f1'].mean()))
print('f1 score (from gridsearch): {}'.format(model.best_score_))
```
